```python
import os, sys
import ujson
from pathlib import Path
from tqdm import tqdm
from collections import defaultdict
from IPython.core.display import display, HTML, Markdown
display(HTML("<style>.container { width:90% !important; }</style>"))
```

```python
def remove_sep_lines(in_file, out_file):
    num_lines = sum(1 for _ in open(in_file))
    with open(in_file, "r") as in_f, open(out_file, "w") as out_f:
        for line in tqdm(in_f, total=num_lines):
            line = ujson.loads(line)
            new_sent = line["sentence"] + ""
            l, r = new_sent.split("[SEP]")
            orig_len_l = len(l.split())
            l = l.strip()
            # If there is a period at the end of the sentence, the spans already account for it
            if l.rstrip(".") != l:
                reduce_offset = 1
            # If not, we are going to add one so the spans are correct (we are swapping . for [SEP])
            else:
                print(line)
                reduct_offset = 0
            new_sent = l.rstrip(".").strip() + " ." + r
            new_spans = []
            for l_s, r_s in line["spans"]:
                if l_s > orig_len_l:
                    l_s -= reduce_offset
                    r_s -= reduce_offset
                new_spans.append([l_s, r_s])
            
            assert new_sent.split(" . ")[0].strip() == line["sentence"].split(" [SEP] ")[0].strip().rstrip(".").strip(), f"{new_sent} VS {line}"
            line["sentence"] = new_sent
            line["spans"] = new_spans
            out_f.write(ujson.dumps(line) + "\n")
    return
```

```python
input_path = Path("/dfs/scratch0/lorr1/projects/bootleg-data/data/benchmarks/aida_0928_nosep/unfiltered")
output_path = Path("/dfs/scratch0/lorr1/projects/bootleg-data/data/benchmarks/aida_0928_nosep/filtered")

os.makedirs(output_path, exist_ok=True)

input_files = input_path.glob("*.jsonl")

for file in input_files:
    output_file = output_path / file.name
    remove_sep_lines(file, output_file)
```
